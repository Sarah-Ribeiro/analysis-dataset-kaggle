### Comparar os funcionários com as empresas

```python
company_location = salaries_data['company_location']
employee_residence = salaries_data['employee_residence']
```

```python
soma_igualdade = sum(company_location == employee_residence)
soma_diferenca = sum(company_location != employee_residence)
```

### Criar novas tabelas para medir o salário 

```python
salary = salaries_data['salary']
salary_in_usd = salaries_data['salary_in_usd']
```

```python
salaries_data['salary_median'] = salary // 2
salaries_data['salary_in_usd_median'] = salary_in_usd // 2
```

```python
print('A média final da coluna salary_in_usd ->',sum(salary_in_usd) // len(salary_in_usd))
# Essa coluna não é muito considerada, já que a moeda está de acordo com o país onde o funcionário vive
print('A média final da coluna salary ->',sum(salary) // len(salary))
```
